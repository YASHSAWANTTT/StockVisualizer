```python

# Import required modules
from kafka import KafkaConsumer
from json import loads
import boto3
from s3fs import S3FileSystem
import json

# Kafka Consumer setup
consumer = KafkaConsumer(
    'stockmarket',  # Replace with your Kafka topic
    bootstrap_servers=['18.219.75.39:9092'],  # Replace with your Kafka broker IP
    value_deserializer=lambda x: loads(x.decode('utf-8'))
)

# AWS S3 setup
s3 = S3FileSystem()

# S3 bucket and folder details
BUCKET_NAME = 'stock-market-analysis-yash'  # Replace with your S3 bucket name
FOLDER_NAME = 'kafka-stock-data'  # Replace with your desired folder name

# Initialize S3 client
s3_client = boto3.client(
    's3',
    aws_access_key_id='AKIAYM7PODJ6AJQC3NNC',  # Replace with your AWS Access Key
    aws_secret_access_key='R28ryqUpWQB5x0UKisiMF8CeLwWu30kUIqypho2A'  # Replace with your AWS Secret Key
)

# Process and store Kafka messages
print("Starting the Kafka Consumer...")
for count, message in enumerate(consumer):
    # Message data
    record = message.value
    print(f"Received message: {record}")
    
    # Define the S3 file path
    file_path = f"{FOLDER_NAME}/record_{count}.json"
    
    # Write the message to S3
    try:
        # Save the JSON file to S3
        with s3.open(f"s3://{BUCKET_NAME}/{file_path}", 'w') as file:
            json.dump(record, file)
        print(f"Saved to S3: s3://{BUCKET_NAME}/{file_path}")
    except Exception as e:
        print(f"Error saving to S3: {e}")
```
